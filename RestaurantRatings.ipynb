```python
object_columns = [c for c in df.columns if df[c].dtypes == 'object']
numeric_columns = [c for c in df.columns if df[c].dtypes != 'object']

sum(df[numeric_columns].isna().sum().to_list())
```

```python
df.drop(object_columns, axis = 1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.25)
```

```python
y = train.Rating.values            # target
X = train.drop(['Rating'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)
```

```python
model = RandomForestRegressor(n_estimators=100, verbose=1, n_jobs=-1, random_state=40)
```

```python
sum(df[numeric_columns].isna().sum().to_list())
```

```python
model.fit(X_train, y_train)
```

```python
def rating_round(x, base=0.5):
    return base * round(x/base)

def predict(ds):
    return np.array([rating_round(x) for x in model.predict(ds)])

y_pred = predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```

```python
plt.rcParams['figure.figsize'] = (10,10)
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(15).plot(kind='barh')
```

```python
predict_submission = predict(test)
```
