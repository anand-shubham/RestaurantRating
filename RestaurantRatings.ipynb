```python
sns.set_style("white")

f, axarr = plt.subplots( )
plt.suptitle('Restaurant Review Metadata', fontsize=18)
f.subplots_adjust(hspace=0.5)

(df
     .groupby('ID_TA')
     .Rating
     .count()
     .sort_values(ascending=False)
     .reset_index(drop=True)
).plot.line()
axarr.set_title("Restaurants by Number of Reviews")

sns.despine()
```

```python
df.Cuisine_Style.unique()
```

```python
df['Cuisine_Qntty'] = df['Cuisine_Style'].str[2:-2].str.split(
        "', '").str.len().fillna(1)
```

```python
df.head(5)
```

```python
df.Reviews.unique()
```

```python
df.Price_Range.unique()
```

```python
price_range = {'$': 0, '$$ - $$$': 1, '$$$$': 2}

df['Pricing'] = df['Price_Range'].map(price_range)
df['Pricing'].fillna(df['Pricing'].mean(), inplace=True)

df['Price_Unknown'] = df['Price_Range'].isna().astype('int32')
    
```

```python
del df['Price_Range']
```

```python
f, axarr = plt.subplots(2,2 ,figsize=(10, 7))
plt.suptitle('Restaurant No. of Cuisines', fontsize=18)
f.subplots_adjust(hspace=0.5)

sns.countplot(df.Pricing.dropna().astype(int),ax=axarr[0][0])
sns.countplot(df.Price_Unknown.dropna().astype(int),ax=axarr[0][1])
sns.countplot(df.Cuisine_Qntty.dropna().astype(int),ax=axarr[1][0])
sns.countplot(df.Rating.dropna().astype(int),ax=axarr[1][1])

sns.despine()
```

```python

corr = df.corr()
#upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

#matplotlip 
f, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},annot=True)
```
